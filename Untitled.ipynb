```python
def Imin_2vr(top):
    
    h=0
    ll=[]
    et=[x+1 for x in A32mBS20BH_ETH]
    Imin=[]
    Imins=[]
    for i in range (4):
        Is=1100

        for ii in top:
            act=utils.bin_series(ACT[ii-1], b=BI)[0]
            act=list(act)
            JDT=np.zeros((BI,4))
            for kk in range (4026-DE):
                aa=act[kk]
                bb=et[kk]
                JDT[aa,bb]= JDT[aa,bb]+1

            JDTP=JDT/(4026-DE)

            oo=JDTP[:,i]
            oo=list(oo)
            xr=0
            XX=[]
            for k in oo:

                if k != 0:
                    h =  (k/sum(oo)) * (math.log2(1/sum(oo))- math.log2(sum(JDTP[xr,:])/k))
                    XX.append(h)

                xr+=1

            s=sum(XX)
            Is= min(Is,s)

        Imins.append(sum(oo)*Is)

    Imin=sum(Imins)
    
    return Imin
```
